# Monte Carlo Estimation of $\pi$

```python
import numpy as np 
from numpy.random import rand
import matplotlib.pyplot as plt
% matplotlib inline
```

#### Array of 'x' and 'y' coordinates

```python
def pi(N):
    x = rand(N)
    y = rand(N)
    z = rand(N)
    
    # Calculate the distance of the points from origin
    distance = (x**2 + y**2 + z**2)**0.5
    
    # Array of points inside the circle
    points_inside = distance[distance < 1]
    
    # Calculate pi by simply taking the ratio of points inside 
    pi = 6*float(len(points_inside))/len(distance)
    return pi
```

```python
plt.scatter(x,y)
plt.show()
```

```python
pi_trial_1 = [pi(10**1), pi(10**2), pi(10**3), pi(10**4), pi(10**5), pi(10**6), pi(10**7), pi(10**8)]
pi_trial_2 = [pi(10**1), pi(10**2), pi(10**3), pi(10**4), pi(10**5), pi(10**6), pi(10**7), pi(10**8)]
pi_trial_3 = [pi(10**1), pi(10**2), pi(10**3), pi(10**4), pi(10**5), pi(10**6), pi(10**7), pi(10**8)]
```

```python
fig, axes = plt.subplots(1,3,figsize=(20,5))
axes[0].plot(pi_trial_1, marker ='o')
axes[0].set_ylim(2.5,3.5)
axes[0].plot([0,7], [3.14159265359,3.14159265359], 'k-', lw=2, alpha =.2)

axes[1].plot(pi_trial_2, marker ='o')
axes[1].set_ylim(2.5,3.5)
axes[1].plot([0,7], [3.14159265359,3.14159265359], 'k-', lw=2, alpha =.2)

axes[2].plot(pi_trial_3, marker ='o')
axes[2].set_ylim(2.5,3.5)
axes[2].plot([0,7], [3.14159265359,3.14159265359], 'k-', lw=2, alpha =.2)

plt.show()
```
